```python
# loading the libraries
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://cytonn.com/blog/article/cytonn-top-employer"

# Make a request to the website.
response = requests.get(url)

# Create a BeautifulSoup object from the response.
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Find the table element that you want to scrape.
table = soup.find("table")
print(table)
```

```python
# creating a list of table rows
rows = table.find_all('tr')
print(rows)
```

```python
# create a list of table columns
columns = []
for row in rows:
    columns.append(row.find_all('td'))
print(columns)
```

```python
import pandas as pd

df = pd.DataFrame(columns, columns=['Rank', 'Company','Rank','Company'])
df.head()
```

```python
# Drop the first row
df = df.drop(df.index[0])

# Print the DataFrame
print(df)
```

```python
df.info()
```

```python
# Convert the 'Rank' and 'Company' columns to a string dtype
df['Rank'] = df['Rank'].astype(str)
df['Company'] = df['Company'].astype(str)
df['Rank'] = df['Rank'].astype(str)
df['Company'] = df['Company'].astype(str)
```

```python
# Rename one of the duplicate columns
df = df.rename(columns={'Rank2': 'Company2'})

# Print the DataFrame
df.head()
```

```python
# Remove the HTML tags
df = df.replace(r"<[^>]*>", "", regex=True)

# Strip whitespace
df = df.applymap(lambda x: x.strip())

# Standardize the case
df = df.applymap(lambda x: x.upper())

# Convert data types
df["Rank"] = df["Rank"].astype(int)
df["Rank"] = df["Rank"].astype(int)

# Remove empty cells
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

# Print the cleaned data
df.head()
```

```python
# export the data to a clean excel file
df.to_csv('company_scraped.csv', index=False)

df.to_excel("company_scraped.xlsx", index=False)
```
